```python
#22BCE0138
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/txhousing.csv"
df = pd.read_csv(url)
print(df)
```

```python
print("\nDataset Dimensions:", df.shape)
```

```python
print("\nSummary of the Dataset:\n", df.describe())
```

```python
print("\nData Types and Missing Values:\n", df.info())
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
#CLeaning the Data by replacing the NA values to their respective mean

numeric_columns = df.select_dtypes(include=[np.number]).columns

df_clean = df.copy()
df_clean[numeric_columns] = df_clean[numeric_columns].fillna(df[numeric_columns].mean())

print("\nMissing Values After Cleaning:\n", df_clean.isnull().sum())
df_clean[['sales', 'volume', 'median', 'listings', 'inventory']] = df_clean[['sales', 'volume', 'median', 'listings', 'inventory']].fillna(df[['sales', 'volume', 'median', 'listings', 'inventory']].mean())

print(pd.isna(df_clean).any())
```

```python
# univariate Analysis

#mean, median and mode of the 'sales', 'volume', 'median', 'listings', 'inventory' column :

for col in numeric_columns:
    mean_value = df[col].mean()
    median_value = df[col].median()
    mode_value = df[col].mode()[0]

    print(f"Column: {col}")
    print(f"Mean: {mean_value}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

features = ['sales', 'volume', 'median', 'listings', 'inventory']
df_clean[features] = df_clean[features].fillna(df_clean[features].mean())

print("\nMissing Values After Cleaning:\n", df_clean[features].isnull().sum())

X = df_clean[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Any NaNs in scaled data:", np.isnan(X_scaled).any())
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

```python
from sklearn.decomposition import PCA

kmeans = KMeans(n_clusters=7, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 7))
plt.scatter(X_pca[y_kmeans == 0, 0], X_pca[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X_pca[y_kmeans == 1, 0], X_pca[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X_pca[y_kmeans == 2, 0], X_pca[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X_pca[y_kmeans == 3, 0], X_pca[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X_pca[y_kmeans == 4, 0], X_pca[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.scatter(X_pca[y_kmeans == 5, 0], X_pca[y_kmeans == 5, 1], s=100, c='orange', label='Cluster 6')
plt.scatter(X_pca[y_kmeans == 6, 0], X_pca[y_kmeans == 6, 1], s=100, c='purple', label='Cluster 7')

centroids_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], s=300, c='yellow', label='Centroids', edgecolor='black')

plt.title('Clusters Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid()
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(X_scaled, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Euclidean Distance')
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
df['sales'] = df['sales'].fillna(df['sales'].mean())
df['volume'] = df['volume'].fillna(df['volume'].mean())
X = df[['sales']]
y = df[['volume']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(f'Intercept: {model.intercept_}')
print(f'Coefficient (Slope): {model.coef_}')
```

```python
y_pred = model.predict(X_test)
evaluate = pd.DataFrame({
    'Actual': y_test.values.flatten(),
    'Predicted': y_pred.flatten()
})
evaluate.head(10)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print(f"R2 score: {r2_score(y_test, y_pred)}")
print(f"MAE score: {mean_absolute_error(y_test, y_pred)}")
print(f"MSE score: {mean_squared_error(y_test, y_pred)}")
```
