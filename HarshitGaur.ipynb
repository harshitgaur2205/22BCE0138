```python
#22BCE0138
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/txhousing.csv"
df = pd.read_csv(url)
print(df)
```

```python
print("\nDataset Dimensions:", df.shape)
```

```python
print("\nSummary of the Dataset:\n", df.describe())
```

```python
print("\nData Types and Missing Values:\n", df.info())
```

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
#CLeaning the Data by replacing the NA values to their respective mean

numeric_columns = df.select_dtypes(include=[np.number]).columns

df_clean = df.copy()
df_clean[numeric_columns] = df_clean[numeric_columns].fillna(df[numeric_columns].mean())

print("\nMissing Values After Cleaning:\n", df_clean.isnull().sum())
df_clean[['sales', 'volume', 'median', 'listings', 'inventory']] = df_clean[['sales', 'volume', 'median', 'listings', 'inventory']].fillna(df[['sales', 'volume', 'median', 'listings', 'inventory']].mean())

print(pd.isna(df_clean).any())
```

```python
# univariate Analysis

#mean, median and mode of the 'sales', 'volume', 'median', 'listings', 'inventory' column :

for col in numeric_columns:
    mean_value = df[col].mean()
    median_value = df[col].median()
    mode_value = df[col].mode()[0]

    print(f"Column: {col}")
    print(f"Mean: {mean_value}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")
```

```python

df_clean = df.dropna(subset=['listings'])

# Plotting the bar chart for 'listings'
plt.figure(figsize=(10,6))
df_clean['listings'].value_counts().nlargest(30).plot(kind='bar')

# Adding title and labels
plt.title("Top 30 Listing Counts")
plt.xlabel('Number of Listings')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Plotting the distribution plot for the 'listings' column
plt.figure(figsize=(10,6))
sns.histplot(df_clean['listings'], bins=30, kde=True)
plt.title("Distribution of Property Listings")
plt.xlabel('Number of Listings')
plt.ylabel('Frequency')
plt.show()
```

```python
# box plot for 'listings' column
plt.figure(figsize=(20,6))
sns.boxplot(x=df_clean['listings'])
plt.title("Boxplot of Property Listings")
plt.xlabel('Number of Listings')

plt.show()
```

```python
# Bivarient Analaysis
df_bivariate = df.dropna(subset=['listings', 'inventory'])
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_bivariate, x='listings', y='inventory')
plt.title("Scatter Plot of Listings vs. Inventory")
plt.xlabel('Number of Listings')
plt.ylabel('Inventory')
plt.show()
```

```python
#box plot for inventory
df_bivariate = df.dropna(subset=['listings', 'inventory']).copy()
df_bivariate.loc[:, 'listings_bins'] = pd.cut(df_bivariate['listings'], bins=5)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_bivariate, x='listings_bins', y='inventory')
plt.title("Box Plot of Inventory Across Listings Bins")
plt.xlabel('Listings Bins')
plt.ylabel('Inventory')
plt.show()
```

```python
# Create a pair plot
df_filtered = df[['sales', 'volume', 'median', 'listings', 'inventory']].dropna()
sns.pairplot(df_filtered, diag_kind='kde', plot_kws={'alpha': 0.5, 's': 50})
plt.suptitle("Pair Plot of Selected Variables", y=1.02)
plt.show()
```

```python
# Calculate the correlation matrix
df_filtered = df[['sales', 'volume', 'median', 'listings', 'inventory']].dropna()
correlation_matrix = df_filtered.corr()
print("Correlation Coefficient Matrix:\n", correlation_matrix)
```

```python
# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# time series analysis of sales as per the dates
time_series_data = df.groupby('date')['sales'].sum().reset_index()
sns.set(style='whitegrid')
plt.figure(figsize=(14, 7))
sns.lineplot(data=time_series_data, x='date', y='sales', marker='o')
plt.title("Total Sales Over Time", fontsize=16)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Total Sales", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/txhousing.csv"
df = pd.read_csv(url)
df['date'] = pd.to_datetime((df['date'] - 1970) * 365.25, unit='D', origin='1970-01-01')
sns.lineplot(data=df, x='date', y='listings', linewidth=0.5, color='blue')
plt.title("Time Series of Listings")
plt.xlabel("Date")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#dots to plot the data for all the other columns:
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns.difference(['rownames', 'date'])

for column in numerical_columns:
    plt.figure(figsize=(16, 6))
    sns.scatterplot(data=df, x='date', y=column, alpha=0.5)
    plt.title(f"Time Series of {column}")
    plt.xlabel("Date")
    plt.ylabel(column)
    plt.xticks(rotation=45)
    plt.grid()
    plt.tight_layout()
    plt.show()
```

```python
#boxplot for grouping time series data
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/txhousing.csv"
df = pd.read_csv(url)
df['date'] = pd.to_datetime((df['date'] - 1970) * 365.25, unit='D', origin='1970-01-01')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='sales', data=df)
plt.title('Boxplot of Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
